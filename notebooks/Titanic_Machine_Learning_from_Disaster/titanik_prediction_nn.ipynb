```python
import pandas as pd
import numpy as np
import keras
```

```python
def write_to_submission_file(predicted_labels, out_file, train_num=891,
                    target='Survived', index_label="PassengerId"):
    # turn predictions into data frame and save as csv file
    predicted_df = pd.DataFrame(predicted_labels,
                                index = np.arange(train_num + 1,
                                                  train_num + 1 +
                                                  predicted_labels.shape[0]),
                                columns=[target])
    predicted_df.to_csv(out_file, index_label=index_label)
```

```python
train_df = pd.read_csv("./data/train.csv") 
test_df = pd.read_csv("./data/test.csv") 
res_df = pd.read_csv("./data/gender_submission.csv")
```

```python
train_df.describe()
train_df.head()
```

```python
test_df.describe()
test_df.head()
```

```python
#cat = keras.utils.to_categorical(train_df['Sex'].astype(int))
cat = pd.get_dummies(train_df['Sex'],  drop_first=True)
train_df['Sex'] = keras.utils.to_categorical(cat)
train_df.head()
```

```python
cat = pd.get_dummies(test_df['Sex'],  drop_first=True)
test_df['Sex'] = keras.utils.to_categorical(cat)
test_df.head()
```

```python
train_df = train_df.drop(['Name','Ticket','Fare','Cabin','Embarked','Age'],axis =1)
test_df = test_df.drop(['Name','Ticket','Fare','Cabin','Embarked','Age'],axis =1)
```

```python
train_df = train_df[['PassengerId','Sex','SibSp','Parch','Pclass','Survived']]
test_df = test_df[['PassengerId','Sex','SibSp','Parch','Pclass']]
```

```python
X_train = train_df.iloc[:,0:5] 
y_train = train_df.iloc[:,5]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(test_df)
```

```python
neural_net = keras.models.Sequential()
neural_net.add(keras.layers.Dense(output_dim = 5, init = 'uniform', activation = 'relu', input_dim = 5))
#Hidden layer
neural_net.add(keras.layers.Dense(output_dim = 4, init = 'uniform', activation = 'relu'))
#output layer with 1 output neuron which will predict 1 or 0
neural_net.add(keras.layers.Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))
```

```python
neural_net.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
neural_net.fit(X_train, y_train, batch_size = 12, nb_epoch = 95)
```

```python
#getting predictions of test data
prediction = neural_net.predict(X_test).tolist()
# list to series
se = pd.Series(prediction)
# creating new column of predictions in data_check dataframe
res_df['check'] = se

res_df['check'] = res_df['check'].str.get(0)
```

```python
series = []
for val in res_df.check:
    if val >= 0.5:
        series.append(1)
    else:
        series.append(0)
res_df['final'] = series
```

```python
np.sum(res_df['final'] != res_df['Survived'])
```

```python
write_to_submission_file(res_df['final'].values, "./data/answer.csv", train_num=891,
                     index_label="PassengerId")
```
