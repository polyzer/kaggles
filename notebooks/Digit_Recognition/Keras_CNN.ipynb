```python
X_train_df = train_df.drop(columns="label")
#X_train_df = standard_scaler.fit_transform(X_train_df)
#X_train_df = keras.utils.np_utils.to_categorical(X_train_df)
X_train_df.shape
```

```python
np.random.seed(7)
# split into 67% for train and 33% for test
X_train, X_valid, y_train, y_valid = train_test_split(X_train_df, Y_train_df, test_size=0.15, random_state=7)
```

```python
X_train = np.reshape(X_train.values, ( X_train.shape[0], 28, 28, 1))
```

```python
#y_train = np.reshape(y_train.values, ( y_train.shape[0], 1))
```

```python
y_train
```

```python
def create_model(train_data):
    # create model
    model = keras.models.Sequential()
    model.add(keras.layers.Conv2D(32, kernel_size=3, activation="relu", input_shape=(28,28,1)))
    model.add(keras.layers.Conv2D(64, kernel_size=3, activation="relu"))
    model.add(keras.layers.Flatten())
    model.add(keras.layers.Dense(10, activation="softmax"))
    # Compile model
    model.compile(optimizer ='sgd', loss = 'sparse_categorical_crossentropy') 
#               metrics =[keras.metrics.mae])
    return model
```

```python
model = create_model(X_train_df)
```

```python
history = model.fit(X_train, y_train, validation_data=(np.reshape(X_valid.values, ( X_valid.shape[0], 28, 28, 1)), np.reshape(y_valid.values, ( y_valid.shape[0], 1))), epochs=30, batch_size=500)
```

```python
X_test_df = test_df
answer = model.predict(np.reshape(X_test_df.values, ( X_test_df.shape[0], 28, 28, 1)))
```

```python
answer = answer.astype(int)
answer
```

```python
y = np.argmax(answer, axis=-1)
y
```

```python
frames_aswer = pd.DataFrame({"ImageId": pd.DataFrame(y).index.values+1,
                             "Label": y
                            })
frames_aswer
```
